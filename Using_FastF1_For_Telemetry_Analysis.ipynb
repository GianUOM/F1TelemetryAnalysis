```python
import fastf1

fastf1.Cache.enable_cache('Cache')
```

```python
year = 2023
gp = "Saudi Arabia"
event = 'R'
session_race = fastf1.get_session(year, gp, event)
session_race.load()
```

```python
session_race.results
```

```python
session_race.laps
```

```python
session_race.laps.columns
```

```python
fastest_lap = session_race.laps.pick_fastest()
fastest_lap
```

```python
driver = 'NOR'
fastest_lap_per = session_race.laps.pick_driver(driver).pick_fastest()
```

```python
fastest_lap_per
```

```python
per_car_data = fastest_lap_per.get_car_data()
```

```python
per_car_data
```

```python
per_car_data.columns
```

```python
from matplotlib import pyplot as plt
from fastf1 import plotting

time = per_car_data['Time']
speed = per_car_data['Speed']
plotting.setup_mpl()
plt.style.use('dark_background')
fig, ax = plt.subplots(figsize=(15,5))
ax.plot(time, speed, label='Fast', color='orange')
ax.grid(True, color='gray', linestyle='--', linewidth=0.5)
ax.set_xlabel('Time')
ax.set_ylabel('Speed(km/hr)')
ax.set_title(driver + "'s Fastest Lap in Speed Telemtry")
plt.legend()
plt.show
```

```python
driver1 = "NOR"
driver2 = "PIA"
driver1_laps = session_race.laps.pick_driver(driver1)
driver2_laps = session_race.laps.pick_driver(driver2)
fig, ax = plt.subplots(figsize=(15,5))
ax.grid(True, color='gray', linestyle='--', linewidth=0.5)
ax.plot(driver1_laps['LapNumber'], driver1_laps['LapTime'], color='red')
ax.plot(driver2_laps['LapNumber'], driver2_laps['LapTime'], color='cyan')
ax.set_title(driver1 + " vs " + driver2)
ax.set_xlabel("Lap Number")
ax.set_ylabel("LapTime")
plt.legend([driver1, driver2])
plt.show()
```

```python
driver1_fastest_lap = session_race.laps.pick_driver(driver1).pick_fastest()
driver2_fastest_lap = session_race.laps.pick_driver(driver2).pick_fastest()
```

```python
driver1_fastest_lap
```

```python
driver2_fastest_lap
```

```python
driver1_car = driver1_fastest_lap.get_car_data()
driver2_car = driver2_fastest_lap.get_car_data()
```

```python
driver1_car
```

```python
plotting.setup_mpl()

fig, ax = plt.subplots(5, figsize = (14,10))

driver1_color = 'red'
driver2_color = 'blue'


ax[0].plot(driver1_car['Time'], driver1_car['Speed'], color=driver1_color)
ax[0].plot(driver2_car['Time'], driver2_car['Speed'], color=driver2_color)
ax[0].set_ylabel("Speed (km/hr)")

ax[1].plot(driver1_car['Time'], driver1_car['RPM'], color=driver1_color)
ax[1].plot(driver2_car['Time'], driver2_car['RPM'], color=driver2_color)
ax[1].set_ylabel("RPM (#)")

ax[2].plot(driver1_car['Time'], driver1_car['nGear'], color=driver1_color)
ax[2].plot(driver2_car['Time'], driver2_car['nGear'], color=driver2_color)
ax[2].set_ylabel("Gear (#)")

ax[3].plot(driver1_car['Time'], driver1_car['Throttle'], color=driver1_color)
ax[3].plot(driver2_car['Time'], driver2_car['Throttle'], color=driver2_color)
ax[3].set_ylabel("Throttle (%)")

ax[4].plot(driver1_car['Time'], driver1_car['Brake'], color=driver1_color)
ax[4].plot(driver2_car['Time'], driver2_car['Brake'], color=driver2_color)
ax[4].set_ylabel("Brake (%)")

ax[0].get_xaxis().set_ticklabels([])
ax[1].get_xaxis().set_ticklabels([])
ax[2].get_xaxis().set_ticklabels([])
ax[3].get_xaxis().set_ticklabels([])
ax[4].get_xaxis().set_ticklabels([])

fig.align_ylabels()
fig.legend([driver1,driver2])
plt.subplots_adjust(left=0.06 ,right=0.99, top=0.9, bottom=0.05)
plt.show()
```
